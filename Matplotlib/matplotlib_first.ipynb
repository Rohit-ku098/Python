## Matplotlib Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Equally spaced 100 numbers between 0 and 10
x = np.linspace(0, 10, 100)

# Sin values for each values in x
y = np.sin(x)

# Cos values for each values in x
z = np.cos(x)
```

### Sin curve plot

```python
plt.plot(x, y)
plt.show()
```

### Cos curve plot

```python
plt.plot(x, z)
plt.show()
```

### Adding title, x-axis label, y-axis label

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("Sine Wave")
3plt.show()
```

### Parabola

```python
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

### Plotting with other symbols

```python
## Plotting curve with + symbol in red color
plt.plot(x, y, "r+")
plt.show()
```

```python
## Plotting curve with . symbol in green color
plt.plot(x, y, "g.")
plt.show()
```

### Multiple curve in single graph

```python
x = np.linspace(-5, 5, 100)

plt.plot(x, np.sin(x), "r-")
plt.plot(x, np.cos(x), "g--")
plt.show()
```

### Bar Graph

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
languages = ["English", "French", "German", "Spanish", "Portuguese"]
peoples = [100, 40, 70, 120, 55]
ax.bar(languages, peoples)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLES")
plt.show()
```

### Pie Chart

```python
fig1 = plt.figure()

ax1 = fig1.add_axes([0, 0, 1, 1])
languages = ["English", "French", "German", "Spanish", "Portuguese"]
peoples = [100, 40, 70, 120, 55]
ax1.pie(peoples, labels=languages, autopct="%1.1f%%")
plt.show()
```

### Scattered Plot

```python
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])

x = np.linspace(0, 10, 50)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x, y, color = "red")
ax.scatter(x, z, color = "green")

plt.show()
```

### 3d scattered plot

```python
fig3 = plt.figure()

ax = plt.axes(projection="3d")

z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x, y, z, c=z, cmap="Blues")
plt.show()
```
