## Advance operations with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000], # Chai Point
])

print("====== Zomato sales analysis ======")
print("Sales Data Shape: ", sales_data.shape)
print("Sample data of 3 row \n", sales_data[:3])
```

```python
# Total Sales Per Year

total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
print("Total Sales Per Year: ", total_sales_per_year)

# Minimum sales per restaurant
min_sales_data = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales per Restaurant: ", min_sales_data)

# Maximum sales per year
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print("Maximum sales per year: ", max_sales_per_year)

# Average sales per restaurant
avg_sales_per_restaurant = np.average(sales_data[:, 1:], axis=1)
print("Average sales per Restaurant: ", avg_sales_per_restaurant)
```

```python
# Cumulative sum
cum_sum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sum : \n", cum_sum)

plt.figure(figsize=(6, 4))
plt.plot(np.mean(cum_sum, axis=0))
plt.title("Average cumulative sum across all restaurants")
plt.xlabel("year")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

## Vector Operations

```python
vector1 = np.array([1,2,3,4])
vector2 = np.array([5,6,7,8])

# vector addition
print("Vector additon: ", vector1+vector1)

#vector multiplication
print("Vector multiplication: ", vector1 * vector2)

# Dot product
print("Dot product: ", np.dot(vector1, vector2))

# Angle -> angle = cosinverse ( a dot b / |a|*|b|)
print("Angle btw vector1 and vector2: ", np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))))
```
